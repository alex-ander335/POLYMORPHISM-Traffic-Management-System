```python
class TrafficDevice:
  """Parent Class representing a generic traffic device."""

  def activate(self):
    raise NotImplementedError("Subclasses must implement the activate method.")


class TrafficLight(TrafficDevice):

  def activate(self):
    print("Traffic Light: Cycling signals (Red/Yellow/Green) to control lane traffic.")


class SpeedCamera(TrafficDevice):

  def activate(self):
    print("Speed Camera: Scanning vehicle speeds and capturing license plates for speeders.")


class PedestrianSignal(TrafficDevice):

  def activate(self):
    print("Pedestrian Signal: Displaying 'WALK' symbol and audio cues for safe crossing.")


class EmergencySiren(TrafficDevice):

  def activate(self):
    print("Emergency Siren: Sounding alarm and clearing traffic lanes for emergency response.")


def main():
  # Store objects in a single list
  devices = [
      TrafficLight(),
      SpeedCamera(),
      PedestrianSignal(),
      EmergencySiren(),  # New device added seamlessly
  ]

  # Polymorphic activation loop - invokes activate() regardless of specific device type
  print("=== Activating Smart Traffic Management System ===")
  for device in devices:
    device.activate()


if __name__ == "__main__":
  main()
```
